This code imports data-science libraries, loads two CSV datasets (train and test) from Google Drive, and merges them into one DataFrame.
It then prints the datasetâ€™s shape and displays the first few rows to quickly inspect the combined data.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")

base_path = "/content/drive/MyDrive/Drug-Response-ML/data/raw"

df_train = pd.read_csv(f"{base_path}/drugsComTrain_raw.csv")
df_test = pd.read_csv(f"{base_path}/drugsComTest_raw.csv")

df = pd.concat([df_train, df_test], axis=0).reset_index(drop=True)

print("Shape:", df.shape)
df.head()
```

```python
df.info()
df.isnull().sum().sort_values(ascending=False)
```

Data Cleaning & Preprocessing

```python
df=df.dropna(subset=['rating'])
df['condition']=df['condition'].fillna('Unknown')
df['review']=df['review'].fillna('')
df['usefulCount']=df['usefulCount'].fillna(0)
df=df.dropna(subset=['drugName'])
df.isnull().sum()
```

```python
# Normalize text to avoid duplicates like "Headache" vs "headache "
df["condition"] = df["condition"].str.strip().str.lower()
df["drugName"] = df["drugName"].str.strip().str.lower()
```

```python
processed_path = "/content/drive/MyDrive/Drug-Response-ML/data/processed/cleaned_data.csv"
df.to_csv(processed_path, index=False)

print("Cleaned dataset saved at:", processed_path)
```
